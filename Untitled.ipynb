```python
# =====================================================
# STEP: K-MODES CLUSTERING (PURE UNSUPERVISED)
# =====================================================

from kmodes.kmodes import KModes
import pandas as pd

# Optional but recommended: drop rows without sector_name
# (sector_name is used ONLY for interpretation)
df_model = df_model.dropna(subset=["sector_name"])


# Categorical features for clustering
k_modes_features = [
    "population_status",
    "category",
    "has_hrp",
    "in_gho"
]

# Prepare K-modes dataset
km_data = df_model[k_modes_features + ["sector_name"]].copy()

# K-modes cannot handle NaN
km_data = km_data.dropna()

km_sample = km_data.sample(
    n=150000,          # you can reduce to 30000 if needed
    random_state=42
)

print("Sample used for K-modes:", km_sample.shape)


# Sanity check
print("K-modes data shape:", km_data.shape)
print(km_data.dtypes)
km_data.head()

# Apply K-modes
km = KModes(
    n_clusters=6,
    init="Huang",
    n_init=3,
    random_state=42,
     verbose=1
)

# km_data["cluster_id"] = km.fit_predict(km_data[k_modes_features])
km_sample["cluster_id"] = km.fit_predict(
    km_sample[k_modes_features]
)


# Cluster sizes
print("\nCluster sizes:")
print(km_sample["cluster_id"].value_counts())

# Sector distribution per cluster (for interpretation)
cluster_sector_distribution = (
    km_sample
    .groupby("cluster_id")["sector_name"]
    .value_counts(normalize=True)
    .rename("proportion")
    .reset_index()
)

cluster_sector_distribution
```

```python
# Cluster size distribution AFTER clustering
after_counts = km_sample["cluster_id"].value_counts().sort_index()

plt.figure()
after_counts.plot(kind="bar")
plt.title("After K-Modes Clustering: Cluster Size Distribution")
plt.xlabel("Cluster ID")
plt.ylabel("Number of Records")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Pivot table for visualization
pivot_sector_cluster = (
    cluster_sector_distribution
    .pivot(index="cluster_id", columns="sector_name", values="proportion")
    .fillna(0)
)

plt.figure()
pivot_sector_cluster.plot(kind="bar", stacked=True)
plt.title("After K-Modes: Sector Composition per Cluster")
plt.xlabel("Cluster ID")
plt.ylabel("Proportion")
plt.tight_layout()
plt.show()
```

```python
target = "sector_name"

features = [
    "population_status",
    "category",
    "population",
    "admin_level",
    "data_year"
]

X = df_model[features]
y = df_model[target]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
categorical_features = X_train.select_dtypes(include="object").columns.tolist()
categorical_features
```

Now I am going to do ordinal encoding “After ordinal encoding, categorical variables are represented numerically, which is required for SMOTE-NC and machine learning models.”

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder(
    handle_unknown="use_encoded_value",
    unknown_value=-1
)

X_train[categorical_features] = encoder.fit_transform(
    X_train[categorical_features]
)

X_test[categorical_features] = encoder.transform(
    X_test[categorical_features]
)


```

```python
# Freeze original encoded training data
X_train_full = X_train.copy()
y_train_full = y_train.copy()
```

```python
X_train[categorical_features].head()
X_train.dtypes

```

```python
cat_feature_indices = [
    X_train.columns.get_loc(col) for col in categorical_features
]

cat_feature_indices
```

```python
X_train_small = X_train_full.sample(n=250000, random_state=42)
y_train_small = y_train_full.loc[X_train_small.index]
```

```python
import matplotlib.pyplot as plt

before_counts = y_train_small.value_counts()

plt.figure()
before_counts.plot(kind="bar")
plt.title("Class Distribution BEFORE SMOTE-NC")
plt.xlabel("Sector Name")
plt.ylabel("Number of Samples")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
from imblearn.over_sampling import SMOTENC

smote_nc = SMOTENC(
    categorical_features=cat_feature_indices,
    sampling_strategy="not majority",   # fast & practical
    random_state=42,
     k_neighbors=1 
)

X_train_res, y_train_res = smote_nc.fit_resample(
    X_train_small, y_train_small
)

```

```python
after_counts = y_train_res.value_counts()

plt.figure()
after_counts.plot(kind="bar")
plt.title("Class Distribution AFTER SMOTE-NC")
plt.xlabel("Sector Name")
plt.ylabel("Number of Samples")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
from catboost import CatBoostClassifier

cat_model = CatBoostClassifier(
    iterations=300,
    learning_rate=0.1,
    depth=8,
    loss_function="MultiClass",
    random_state=42,
    verbose=100
)

cat_model.fit(
    X_train_res,
    y_train_res
)
```

```python
y_pred_cat = cat_model.predict(X_test)
y_pred_cat = y_pred_cat.flatten()
```

```python
from sklearn.metrics import classification_report, accuracy_score

print("CatBoost Accuracy:", accuracy_score(y_test, y_pred_cat))
print(classification_report(y_test, y_pred_cat))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_cat, digits=3))
```

```python
from sklearn.metrics import top_k_accuracy_score

y_proba = cat_model.predict_proba(X_test)
top3_acc = top_k_accuracy_score(
    y_test,
    y_proba,
    k=3,
    labels=cat_model.classes_
)

print("Top-3 Accuracy:", top3_acc)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4, 4))
plt.bar(["Top-1", "Top-3"], [accuracy_score(y_test, y_pred_cat), top3_acc])
plt.ylim(0, 1)
plt.title("CatBoost Performance Summary")
plt.ylabel("Accuracy")
plt.tight_layout()
plt.show()
```
